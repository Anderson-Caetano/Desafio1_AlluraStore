### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Inicializando um dicionário para armazenar o faturamento de cada loja
faturamento_lojas = {}
# Define the urls variable as a list containing the URLs of the four stores
urls = [url, url2, url3, url4]

# Loop para calcular o faturamento de cada loja
for i, url in enumerate(urls, start=1):
    try:
        # Carregar os dados da loja
        df = pd.read_csv(url)
        # Calcular o faturamento
        faturamento = df['Preço'].sum()
        # Armazenar o faturamento
        faturamento_lojas[f'Loja_{i}'] = faturamento
    except Exception as e:
        print(f"Erro ao processar a Loja {i}: {e}")

# Exibindo o faturamento total de cada loja
for loja, faturamento in faturamento_lojas.items():
    print(f"{loja}: R$ {faturamento:.2f}")
```

# 6. Gráficos

```python
# Inicializando um dicionário para armazenar o faturamento de cada loja
import pandas as pd
import matplotlib.pyplot as plt

faturamento_lojas = {}
# Define the urls variable as a list containing the URLs of the four stores
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
urls = [url, url2, url3, url4]

# Inicializando listas para gráficos
faturamentos = []
lojas = []
fretes_medios = []
avaliacoes_medias = []

# Loop para calcular o faturamento de cada loja
for i, url in enumerate(urls, start=1):
    try:
        # Carregar os dados da loja
        df = pd.read_csv(url)

        # Calcular o faturamento
        faturamento = df['Preço'].sum()
        faturamento_lojas[f'Loja_{i}'] = faturamento
        faturamentos.append(faturamento)
        lojas.append(f'Loja_{i}')

        # Contar a quantidade de produtos vendidos por categoria
        categoria_counts = df['Categoria do Produto'].value_counts()
        print(f"Loja_{i} - Quantidade de produtos vendidos por categoria:\n{categoria_counts}\n")

        # Calcular a média das avaliações dos clientes
        avaliacao_media = df['Avaliação da compra'].mean()
        avaliacoes_medias.append(avaliacao_media)
        print(f"Loja_{i} - Média das avaliações dos clientes: {avaliacao_media:.2f}\n")

        # Identificar os produtos mais vendidos e menos vendidos
        produto_counts = df['Produto'].value_counts()
        mais_vendidos = produto_counts.head(5)
        menos_vendidos = produto_counts.tail(5)

        print(f"Loja_{i} - Produtos mais vendidos:\n{mais_vendidos}\n")
        print(f"Loja_{i} - Produtos menos vendidos:\n{menos_vendidos}\n")

        # Calcular o custo médio de frete
        custo_medio_frete = df['Frete'].mean()
        fretes_medios.append(custo_medio_frete)
        print(f"Loja_{i} - Custo médio de frete: R$ {custo_medio_frete:.2f}\n")

    except Exception as e:
        print(f"Erro ao processar a Loja {i}: {e}")

# Exibindo o faturamento total de cada loja
for loja, faturamento in faturamento_lojas.items():
    print(f"{loja}: R$ {faturamento:.2f}")

# Gráfico 1: Faturamento por loja
plt.figure(figsize=(10, 6))
plt.bar(lojas, faturamentos, color='blue')
plt.title('Faturamento por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico 2: Custo médio de frete por loja
plt.figure(figsize=(10, 6))
plt.bar(lojas, fretes_medios, color='green')
plt.title('Custo Médio de Frete por Loja')
plt.xlabel('Lojas')
plt.ylabel('Custo Médio de Frete (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico 3: Média das avaliações dos clientes por loja
plt.figure(figsize=(10, 6))
plt.plot(lojas, avaliacoes_medias, marker='o', linestyle='-', color='red')
plt.title('Média das Avaliações dos Clientes por Loja')
plt.xlabel('Lojas')
plt.ylabel('Média das Avaliações')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 7. Análise de Desempenho Geográfico

```python
# Inicializando um dicionário para armazenar o faturamento de cada loja
import matplotlib.pyplot as plt
import folium

faturamento_lojas = {}
# Define the urls variable as a list containing the URLs of the four stores
urls = [url, url2, url3, url4]

# Inicializando listas para gráficos
faturamentos = []
lojas = []
fretes_medios = []
avaliacoes_medias = []

# Inicializando lista para armazenar coordenadas
coordenadas = []

# Loop para calcular o faturamento de cada loja
for i, url in enumerate(urls, start=1):
    try:
        # Carregar os dados da loja
        df = pd.read_csv(url)

        # Calcular o faturamento
        faturamento = df['Preço'].sum()
        faturamento_lojas[f'Loja_{i}'] = faturamento
        faturamentos.append(faturamento)
        lojas.append(f'Loja_{i}')

        # Armazenar coordenadas
        coordenadas.extend(zip(df['lat'], df['lon']))

        # Contar a quantidade de produtos vendidos por categoria
        categoria_counts = df['Categoria do Produto'].value_counts()
        print(f"Loja_{i} - Quantidade de produtos vendidos por categoria:\n{categoria_counts}\n")

        # Calcular a média das avaliações dos clientes
        avaliacao_media = df['Avaliação da compra'].mean()
        avaliacoes_medias.append(avaliacao_media)
        print(f"Loja_{i} - Média das avaliações dos clientes: {avaliacao_media:.2f}\n")

        # Identificar os produtos mais vendidos e menos vendidos
        produto_counts = df['Produto'].value_counts()
        mais_vendidos = produto_counts.head(5)
        menos_vendidos = produto_counts.tail(5)

        print(f"Loja_{i} - Produtos mais vendidos:\n{mais_vendidos}\n")
        print(f"Loja_{i} - Produtos menos vendidos:\n{menos_vendidos}\n")

        # Calcular o custo médio de frete
        custo_medio_frete = df['Frete'].mean()
        fretes_medios.append(custo_medio_frete)
        print(f"Loja_{i} - Custo médio de frete: R$ {custo_medio_frete:.2f}\n")

    except Exception as e:
        print(f"Erro ao processar a Loja {i}: {e}")

# Exibindo o faturamento total de cada loja
for loja, faturamento in faturamento_lojas.items():
    print(f"{loja}: R$ {faturamento:.2f}")

# Gráfico 1: Faturamento por loja
plt.figure(figsize=(10, 6))
plt.bar(lojas, faturamentos, color='blue')
plt.title('Faturamento por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico 2: Custo médio de frete por loja
plt.figure(figsize=(10, 6))
plt.bar(lojas, fretes_medios, color='green')
plt.title('Custo Médio de Frete por Loja')
plt.xlabel('Lojas')
plt.ylabel('Custo Médio de Frete (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico 3: Média das avaliações dos clientes por loja
plt.figure(figsize=(10, 6))
plt.plot(lojas, avaliacoes_medias, marker='o', linestyle='-', color='red')
plt.title('Média das Avaliações dos Clientes por Loja')
plt.xlabel('Lojas')
plt.ylabel('Média das Avaliações')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Mapa de calor das vendas por localização
mapa = folium.Map(location=[-15.7801, -47.9292], zoom_start=4)
from folium.plugins import HeatMap

# Mapa de calor das vendas por localização
mapa = folium.Map(location=[-15.7801, -47.9292], zoom_start=4)

# Adicionar Heatmap
HeatMap(coordenadas, radius=10, blur=15, max_zoom=10).add_to(mapa)

# Salvar o mapa como HTML
mapa.save('mapa_vendas.html')
import webbrowser
webbrowser.open('mapa_vendas.html')
print("Mapa de calor salvo como 'mapa_vendas.html'")
```
